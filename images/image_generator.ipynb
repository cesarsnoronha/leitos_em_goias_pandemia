Abaixo, histogramas que n√£o foram utilizados no relat√≥rio

```python
from scipy import stats
import matplotlib.pyplot as plt

# assume you have two dataframes df1 and df2
df1= df[df['date'] == '2020-03'] #2020
df2= df[df['date'] == '2022-03'] #2022

# calculate the z-scores for QT_EXIST
z_scores_exist = stats.zscore(df1['QT_EXIST0'])
df1 = df1[(z_scores_exist < 3)]

# calculate the z-scores for QT_SUS
z_scores_sus = stats.zscore(df2['QT_SUS'])
df2 = df2[(z_scores_sus < 3)]

# create the first subplot
plt.subplot(1, 2, 1)
plt.hist(df1['QT_EXIST0'], bins=40)
plt.hist(df1['QT_SUS'], bins=40)
plt.title('QT_EXIST0')

# create the second subplot
plt.subplot(1, 2, 2)
plt.hist(df2['QT_EXIST0'], bins=40)
plt.hist(df2['QT_SUS'], bins=40)
plt.title('QT_SUS')

# display the subplot
plt.show()

```

```python
# Per√≠odo de redu√ß√£o 1
df_reducao_11 = df[df.date == '2020-10'].groupby('CODUFMUN')['QT_EXIST0'].sum().reset_index()
df_reducao_12 = df[df.date == '2020-11'].groupby('CODUFMUN')['QT_EXIST0'].sum().reset_index()

# Per√≠odo de redu√ß√£o 2
df_reducao_21 = df[df.date == '2021-12'].groupby('CODUFMUN')['QT_EXIST0'].sum().reset_index()
df_reducao_22 = df[df.date == '2022-06'].groupby('CODUFMUN')['QT_EXIST0'].sum().reset_index()
```
